## PCA:
Depois de formamos nossa matriz e realizarmos o fit_transform dos seus dados, faremos o pca que ira comprimir e agrupar os dados em dois componentes. O PCA é algo muito interessante de se aplicar aos dados, pois ele ira agrupar seus dados em componentes levando em conta os dados que são passados. Ele dará pesos e analisará qual coisa ficará em qual grupo de componente.

````python
finalDf = pickle.load( open( "pca_texto.p", "rb" ) )
finalDf.head(10)
````

## Word Cloud:
Com a nuvem de palavras, podemos ter um visão mais clara das palavras que mais se repetem no texto que foi compilado e com isso, podemos ter uma ideia, um conhecimento prévio do que todas aquelas páginas tratam. 

````python
texto = pickle.load( open( "textos.p", "rb" ) )

# Carrega as stopwords e adiciona outras palavras, simbolos e números para retirar
all_stopwords = stopwords.words('portuguese')
all_stopwords += punctuation 
all_stopwords += ['','``','a','em','o','e'] 
all_stopwords += [str(i) for i in range(0,1000)]

# Criando a wordcloud
wordcloud = WordCloud(stopwords=all_stopwords,
                      background_color='black', width=1600,                            
                      height=800).generate(texto)


# Realizando a plotagem 
fig, ax = plt.subplots(figsize=(16,8))            
ax.imshow(wordcloud, interpolation='bilinear')       
ax.set_axis_off()
plt.imshow(wordcloud)
plt.show()               
````
